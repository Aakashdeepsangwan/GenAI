```python
from docx import Document  as Docx
import os
```

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
# Using Docx2txtLoader

try :
    docs_loader = Docx2txtLoader("/Users/akashdeepsangwan/Desktop/Code/Projects/GenAI/RAG/Data_Ingestion_Parsing/data/text_files/AI & Automation Data Scientist.docx")
    docs = docs_loader.load()

    print(f"Loaded {len(docs)} documents")
    print(f"Content PreViewed : {docs[0].page_content[:200]}....")
    
    print(    )
    print(f"Metadata : {docs[0].metadata}")

except Exception as e :
    print(f"Error : {e}")
```

```python
# Method 2

print("Using unstructuredWordDocumentLoader")

try:
    # mode = structure of the document, there is way, we will be reading the files
    unstructured_loader = UnstructuredWordDocumentLoader("/Users/akashdeepsangwan/Desktop/Code/Projects/GenAI/RAG/Data_Ingestion_Parsing/data/text_files/AI & Automation Data Scientist.docx", mode = "single")
    loading_file= unstructured_loader.load()

    print(f"Loader {len(loading_file)} elements")

    for i, doc in enumerate(loading_file[:3]):
        print(f"Element : {i+1} : ")
        print(f"Type : {doc.metadata.get('category', 'unknown')}")
        print(f"Content : {doc.page_content[:200]}...")

except Exception as e :
    print(f"Error : {e}")
```
